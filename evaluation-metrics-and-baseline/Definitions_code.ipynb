# Portfolio Evaluation Metrics — Quick Reference Pamphlet

This is a simple, clear overview of the key metrics I’ll use to evaluate portfolio performance.  
Definitions, typical ranges, and quick examples are included so I can interpret results instantly.

---

## **1. Total Return**
**What it measures:**  
How much the portfolio gained or lost from the start to the end.

**Formula:**  
```
(V_end - V_start) / V_start
```

**Interpretation:**  
- Positive → grew  
- Negative → lost  
- Higher is better  

**Typical ranges:**  
- **5–15% yearly** → average  
- **15–30%** → strong  
- **30%+** → exceptional  

**Example:**  
Start = 100 → End = 118  
Total return = **18%**

---

## **2. Sharpe Ratio**
**What it measures:**  
Risk-adjusted return. How much return I’m getting per unit of volatility.

**Formula:**  
```
(mean_return - risk_free_rate) / std_return
```

**Interpretation:**  
Higher is better.

**Typical ranges:**  
- **0.0 – 1.0** → weak  
- **1.0 – 1.5** → acceptable  
- **1.5 – 2.0** → strong  
- **2.0+** → hedge-fund quality  

**Example:**  
Mean monthly return = 1.2%  
Monthly Volatility = 3%  
Sharpe = **0.40**

---

## **3. Maximum Drawdown**
**What it measures:**  
Largest drop from a peak to the next trough.  
Shows the worst “pain period” of the strategy.

**Interpretation:**  
- Small drawdowns = safer  
- Large drawdowns = emotionally hard + riskier  

**Typical ranges:**  
- **–5% to –15%** → low-risk strategies  
- **–15% to –30%** → typical equity portfolios  
- **–30% to –60%** → high-volatility strategies  

**Example:**  
Peak = 110 → Trough = 90  
Max drawdown = **–18.2%**

---

## **4. Volatility (Annualized)**
**What it measures:**  
How much periodic returns fluctuate.

**Formula:**  
```
std_returns * sqrt(periods_per_year)
```

**Interpretation:**  
Lower volatility = smoother ride.

**Typical ranges:**  
- **5–10%** → very stable  
- **10–20%** → typical equities  
- **20–40%** → aggressive strategies  
- **40%+** → very risky  

**Example:**  
Monthly std = 3%  
Annualized volatility = 0.03 × √12 = **10.4%**

---

## **5. Worst Losing Streak**
**What it measures:**  
Longest run of consecutive negative return periods.

**Interpretation:**  
Important for psychology and sticking with the strategy.

**Typical ranges:**  
- **1–3 months** → stable strategies  
- **4–8 months** → common in equities  
- **9+ months** → hard to hold  

**Example:**  
Returns contain 3 negative months in a row → Losing streak = **3**

---

## **6. Turnover**
**What it measures:**  
How much the portfolio composition changes at each rebalance.

**Formula:**  
```
sum( |new_weight - old_weight| ) / 2
```

**Interpretation:**  
- Low turnover = cheaper, more stable  
- High turnover = more trading costs, more noise  

**Typical ranges:**  
- **0–10% quarterly** → very low  
- **10–30%** → normal  
- **30–60%** → aggressive  
- **60%+** → high-frequency style  

**Example:**  
Old = [0.2, 0.3, 0.5]  
New = [0.3, 0.3, 0.4]  
Turnover = **0.10 (10%)**
