This notebook contains clear, correct, professional definitions of the core metrics used to evaluate portfolio performance. These will serve as the foundation for your LSTM-based stock selection strategy and future hedge-fund-level development.

---

## **1. Total Return**
**Definition:**  
The percentage change in portfolio value from start to end.

$$
\text{Total Return} = \frac{V_{end} - V_{start}}{V_{start}}
$$

---

## **2. Sharpe Ratio**
**Definition:**  
Risk-adjusted return. Measures excess return per unit of volatility.

$$
\text{Sharpe} = \frac{R_p - R_f}{\sigma_p}
$$

Where:  
- \(R_p\): average portfolio return  
- \(R_f\): risk-free rate (can be set to 0 for simplicity)  
- \(\sigma_p\): standard deviation of returns  

---

## **3. Maximum Drawdown**
**Definition:**  
Worst peak-to-trough drop in portfolio value.

$$
\text{Max Drawdown} =
\max_t \left( \frac{Peak(t) - Value(t)}{Peak(t)} \right)
$$

---

## **4. Volatility**
**Definition:**  
Standard deviation of returns, annualized.

$$
\sigma_{annual} = \sigma_{period} \times \sqrt{N}
$$

Where \(N\) is number of periods per year (12 months, 52 weeks, 252 days).

---

## **5. Worst Losing Streak**
**Definition:**  
Longest sequence of consecutive periods with negative returns.

---

## **6. Turnover**
**Definition:**  
How much the portfolio composition changes during rebalancing.

$$
\text{Turnover} = \frac{\sum_i |w_i^{new} - w_i^{old}|}{2}
$$

This captures trading intensity and cost.
